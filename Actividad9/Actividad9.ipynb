## **Ejercicio 1.** 

\begin{eqnarray*}
\frac{d x}{d t} & = & y  \\
\frac{d y}{dt } & = & -x   \\
\end{eqnarray*}

\begin{eqnarray*}
\frac{d x}{d t} & = & 0 x + 1y \\
\frac{d y}{dt } & = & -1x + 0y  \\
\end{eqnarray*}

\begin{equation*}
\frac{d}{dt} \begin{bmatrix}
x \\ y \end{bmatrix} = A \begin{bmatrix}
x \\ y \end{bmatrix}, \mathrm{donde} \ A = \begin{bmatrix}
0 & 1 \\ -1 & 0 \end{bmatrix}
\end{equation*}

```
import numpy as np
from scipy import integrate, linalg
import pylab as plt

# Definimos un vector X: X[0]=x, X[1]=y
# Definimos la Matriz A 

A1 = np.array([[0,1], [-1,0]])
print('A =', A1)

def dX_dt_1(X, t=1):
    """ El lado derecho del sistema de ecuaciones """
    return np.array([ 0*X[0] + X[1] , -1*X[0] + 0*X[1]])
    
# Los puntos críticos del sistema
X_f0 = np.array([     0. ,  0.])

print('Punto crítico del sistema')
print('Punto crítico X_f0:', X_f0)
print(' ')
# En los puntos críticos se anulan las derivadas. 
all(dX_dt_1(X_f0) == np.zeros(2) )
                                
# Aplicamos la biblioteca scipy.linalg.eig()
eigvals1, eigvecs1 = linalg.eig(A1)
print('Eigenvalores:', eigvals1)
print('Eigenvectores:', eigvecs1)
```

```
# Solución
# Los eigenvectores son linealmente independientes
lambda1_1 = eigvals1[0]
lambda2_1 = eigvals1[1]
print('Eigenvalores')
print('lambda1= ', lambda1_1)
print('lambda2= ', lambda2_1)
print(' ')
print('Eigenvectores')
v1_1 = eigvecs1[:,0]
v2_1 = eigvecs1[:,1]
print('v1=',v1_1)
print('v2=',v2_1)
```

```
# Solución numérica
# Integramos el sistema de ecuaciones de Lotka-Volterra usando odeint

t = np.linspace(-50, 50,  500)              # Rango de tiempo de integración
i = 1
while i <= 5:
  X0_1 = np.array([i, 0])                     # Condiciones iniciales
  X, infodict = integrate.odeint(dX_dt_1, X0_1, t, full_output=True)
  infodict['message']
  x,y = X.T
  plt.plot(x, y)

  i = i + 1  

plt.grid()
plt.title('Solución X(t) en el plano fase')
plt.xlabel('x')
plt.ylabel('y')
plt.figure(figsize=(5,5))
plt.show()
```

## **Ejercicio 3.** 

\begin{eqnarray*}
\frac{d^2 x}{d t^2} + \omega_0^2 x & = & 0, \ \omega_0 > 0  \\
\end{eqnarray*}

Empezamos a despejar el diferencial:

\begin{eqnarray*}
\frac{d^2 x}{d t^2} = -\omega_0^2 x & \\
u = \frac{dx}{dt} \\
\frac{du}{dt} = -\omega_0^2 x
\end{eqnarray*}

Nos queda:

\begin{eqnarray*}
\frac{du}{dt} & = & -\omega_0^2 x   + 0u \\
\frac{dx}{dt} & = & 0x + 1u  \\
\end{eqnarray*}

Reescribimos:

\begin{equation*}
\frac{d}{dt} \begin{bmatrix}
u \\ x \end{bmatrix} = A \begin{bmatrix}
u \\ x \end{bmatrix}, \mathrm{donde} \ A = \begin{bmatrix}
-\omega_0^2 & 0 \\ 0 & 1 \end{bmatrix}
\end{equation*}

```
import numpy as np
from scipy import integrate, linalg
import pylab as plt

# Definimos un vector X: X[0]=x, X[1]=y
# Definimos la Matriz A 

A3 = np.array([[-1,0], [0,1]])
print('A =', A3)

def dX_dt_3(X, t=1):
    """ El lado derecho del sistema de ecuaciones """
    return np.array([ -1*X[0] + 0*X[1] , 0*X[0] + 1*X[1]]) # w0 = 1
    
# Los puntos críticos del sistema
X_f0 = np.array([     0. ,  0.])

print('Punto crítico del sistema')
print('Punto crítico X_f0:', X_f0)
print(' ')
# En los puntos críticos se anulan las derivadas. 
all(dX_dt_3(X_f0) == np.zeros(2) )
                                
# Aplicamos la biblioteca scipy.linalg.eig()
eigvals3, eigvecs3 = linalg.eig(A3)
print('Eigenvalores:', eigvals3)
print('Eigenvectores:', eigvecs3)
```

## **Ejercicio 4.** 

\begin{eqnarray*}
\frac{dx}{dt} & = & -2 x  \\
\frac{dy}{dt} & = &  2 z  \\
\frac{dz}{dt} & = &  - 2 y  \\
\end{eqnarray*}

Reescribimos:

\begin{eqnarray*}
\frac{dx}{dt} & = & -2x + 0y + 0z \\
\frac{dy}{dt} & = & 0x + 0y + 2z  \\
\frac{dz}{dt} & = & 0x - 2y + 0z  \\
\end{eqnarray*}

\begin{equation*}
\frac{d}{dt} \begin{bmatrix}
x \\ y \\ z \end{bmatrix} = A \begin{bmatrix}
x \\ y \\ z \end{bmatrix}, \mathrm{donde} \ A = \begin{bmatrix}
-2 & 0 & 0 \\ 0 & 0 & 2 \\ 0 & -2 & 0 \end{bmatrix}
\end{equation*}

```
import numpy as np
from scipy import integrate, linalg
import pylab as plt

# Definimos un vector X: X[0]=x, X[1]=y, X[2]=z
# Definimos la Matriz A 

A4 = np.array([[-2,0,0],[0,0,2], [0,-2,0]])
print('A =', A4)

def dX_dt_4(X,t=0):
    """ El lado derecho del sistema de ecuaciones """
    return np.array([-2*X[0] + 0*X[1] + 0*X[2], 0*X[0] + 0*X[1] + 2*X[2] , 0*X[0] - 2*X[1] + 0*X[2]])
    
# Los puntos críticos del sistema
X_f0 = np.array([ 0.0 , 0.0, 0.0])

print('Punto crítico del sistema')
print('Punto crítico X_f0:', X_f0)
print(' ')
# En los puntos críticos se anulan las derivadas. 
all(dX_dt_4(X_f0) == np.zeros(3) )
                                
# Aplicamos la biblioteca scipy.linalg.eig()
eigvals4, eigvecs4 = linalg.eig(A4)
print('Eigenvalores:', eigvals4)
print('Eigenvectores:', eigvecs4)
```

```
# Solución
# Los eigenvectores son linealmente independientes
lambda1_4 = eigvals4[0]
lambda2_4 = eigvals4[1]
lambda3_4 = eigvals4[2]
print('Eigenvalores')
print('lambda1= ', lambda1_4)
print('lambda2= ', lambda2_4)
print('lambda3= ', lambda3_4)
print(' ')
print('Eigenvectores')
v1_4 = eigvecs4[:,0]
v2_4 = eigvecs4[:,1]
v3_4 = eigvecs4[:,2]
print('v1=',v1_4)
print('v2=',v2_4)
print('v3=',v3_4)
```

```
# Solución numérica
# Integramos el sistema de ecuaciones de Lotka-Volterra usando odeint

t = np.linspace(-50, 50,  500)              # Rango de tiempo de integración
i = 1
k = 1
ax = plt.axes(projection ='3d')
plt.xlabel('x')
plt.ylabel('y')

while i <= 5 and k <= 5:
  X0_4 = np.array([i, 1, k])                     # Condiciones iniciales
  X, infodict = integrate.odeint(dX_dt_4, X0_4, t, full_output=True)
  infodict['message']
  x,y,z = X.T
  ax.plot3D(x, y, z)

  i = i + 1  
  k = k + 1

fig = plt.figure()
ax.set_title('Solución en el espacio fase (x(t),y(t),z(t))')
plt.show()
```

```
import numpy as np
from scipy import integrate, linalg
import pylab as plt

# Definimos un vector X: X[0]=x, X[1]=y, X[2]=z
# Definimos la Matriz A 

A9 = np.array([[2,0,0],[0,-1,0], [0,2,-3]])
print('A =', A9)

def dX_dt_9(X,t=0):
    """ El lado derecho del sistema de ecuaciones """
    return np.array([2*X[0] + 0*X[1] + 0*X[2], 0*X[0] - 1*X[1] + 0*X[2] ,0*X[0] + 2*X[1] - 3*X[2]])
    
# Los puntos críticos del sistema
X_f0 = np.array([ 0.0 , 0.0, 0.0])

print('Punto crítico del sistema')
print('Punto crítico X_f0:', X_f0)
print(' ')
# En los puntos críticos se anulan las derivadas. 
all(dX_dt_9(X_f0) == np.zeros(3) )
                                
# Aplicamos la biblioteca scipy.linalg.eig()
eigvals9, eigvecs9 = linalg.eig(A9)
print('Eigenvalores:', eigvals9)
print('Eigenvectores:', eigvecs9)
```

El sistema esta sujeto del primer resorte de la izquierda. Cuando el sistema se deja en reposo, las longitudes de los resortes son $L_1$ y $L_2$.

Las ecuaciones de movimiento estan dadas por las siguientes ecuaciones diferenciales acopladas:

\begin{eqnarray*}
m_1 x_1^{''} + b_1 x_1^{'}+k_1 (x_1-L_1)-k_2(x_2-x_1-L_2) & = & 0 \\

\begin{eqnarray*}
x_1^{''} + b_1 x_1^{'}+k_1 (x_1-1)-k_2(x_2-x_1-1) & = & 0 \\
x_2^{''} + b_2 x_2^{'}+k_2(x_2-x_1-1) & = & 0 \\
\end{eqnarray*}

Simplificamos

\begin{eqnarray*}
x_1^{''} = - b_1 x_1^{'} - k_1x_1 + k_1 + k_2(x_2-x_1) - k_2 & = & 0 \\

\end{eqnarray*}

Cambio de variable
\begin{eqnarray*}
x_1^{'} = x_3 \\

Sustituimos:
